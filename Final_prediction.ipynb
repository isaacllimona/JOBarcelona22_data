```python
import joblib
import pandas as pd
```

```python
test_data_path = "test_x.csv"
test_data = pd.read_csv(test_data_path)
test_data
```

Load the classifier and make the predictions

```python
classifier = joblib.load("classifier.joblib")
```

```python
y = classifier.predict(test_data)
```

save predictions to file

```python
submission_data = pd.DataFrame(y)
submission_data.to_csv("results.csv", index=True)
```
